```python
#importing all the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading CSV 
df = pd.read_csv("pubg.csv")
df
```

```python
#describe():
nc_summary = df.describe()
nc_summary
```

```python
avgper = df['kills'].mean()
print("\nThe average person kills :", avgper,"players")
```

```python
ppl = df["kills"].quantile(0.99)
print("\n99% of people have",ppl,"kills")
```

```python
most_kills = df["kills"].max()
print("\nThe most kills ever recorded are :",most_kills)
```
